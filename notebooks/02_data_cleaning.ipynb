# Data Cleaning and Feature Engineering

This notebook standardizes the retail dataset into an analytics-ready table that can be reused across EDA, RFM segmentation, churn analysis, forecasting, Power BI, and the website dashboard.


```python
from pathlib import Path
import sys

import pandas as pd
from IPython.display import display

pd.set_option("display.max_columns", None)

project_root = Path.cwd()
if not (project_root / "data").exists() and (project_root.parent / "data").exists():
    project_root = project_root.parent

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from scripts.data_utils import clean_superstore_data, read_csv_with_fallback

raw_path = project_root / "data" / "raw" / "superstore_raw.csv"
cleaned_path = project_root / "data" / "cleaned" / "superstore_cleaned.csv"

raw_df, used_encoding = read_csv_with_fallback(raw_path)
print(f"Loaded raw data with encoding: {used_encoding}")
```

## Clean and Standardize the Dataset


```python
cleaned_df = clean_superstore_data(raw_df)

print(f"Cleaned shape: {cleaned_df.shape[0]:,} rows x {cleaned_df.shape[1]:,} columns")
cleaned_df.head()
```

## Why These Cleaning Steps Matter

- Standardized column names make code reusable and reduce transformation errors.
- Removing corrupt or utility fields prevents noise from leaking into business analysis.
- Converting dates correctly is essential for recency, seasonality, and forecast modeling.
- Time features support trend analysis, cohort slicing, and Power BI filtering.
- Profit margin helps evaluate business quality, not just top-line volume.


```python
cleaned_df.info()
```

## Missing Values and Duplicates After Cleaning


```python
missing_post_cleaning = (
    pd.DataFrame(
        {
            "missing_count": cleaned_df.isna().sum(),
            "missing_percent": (cleaned_df.isna().mean() * 100).round(2),
        }
    )
    .sort_values("missing_count", ascending=False)
)
duplicate_post_cleaning = cleaned_df.duplicated().sum()

display(missing_post_cleaning[missing_post_cleaning["missing_count"] > 0])
print(f"Duplicate rows after cleaning: {duplicate_post_cleaning:,}")
```

## Save Cleaned Dataset


```python
cleaned_path.parent.mkdir(parents=True, exist_ok=True)
cleaned_df.to_csv(cleaned_path, index=False)
print(f"Saved cleaned dataset to: {cleaned_path.resolve()}")
```

## Notebook Outcome

The dataset is now standardized, enriched with time fields and margin logic, and ready for downstream analytics.
